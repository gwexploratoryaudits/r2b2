```python
import pandas as pd

fname = 'Virginia_Elections_Database__2016_President_General_Election_including_precincts.csv'
df = pd.read_csv(fname)
df
```

```python
# create a dictionary containing a list of all ballots
# where each ballot has id number 1...N, candidate, and precinct
from itertools import count


cands = ['Hillary R. Clinton', 'Donald J. Trump', 'Gary Johnson', 'Evan McMullin', 'Jill Stein', 'All Others']
bals = {"bals": []}
precinct_list = {"precinct_list": []}
county_list = {"county_list": []}
index = 0
county_and_precinct_info = {}
for precIndex in range(1,2749):
    precinctRow = df.iloc[precIndex]
    for cand in cands:
        votesForCand = int(precinctRow[cand].replace(',', ''))
        for i in range(votesForCand):
            bals["bals"].append({
                'id': index,
                'precinct': precinctRow['Pct'],
                'county': precinctRow['County/City'],
                'candidate': cand
            })
            county = precinctRow['County/City']
            if county not in county_and_precinct_info:
                county_and_precinct_info.update({county:[]})
            precinct = precinctRow['Pct']
            if precinct not in county_and_precinct_info[county]:
                county_and_precinct_info[county].append(precinct)
            precinct_list["precinct_list"].append(precinctRow['Pct'])
            county_list["county_list"].append(precinctRow['County/City'])
            index += 1
precinct_list["precinct_list"] = list(set(precinct_list["precinct_list"]))
county_list["county_list"] = list(set(county_list["county_list"]))
```

```python
county_list
```

```python

per_county_votes = {}
for county in county_list["county_list"]:
    per_county_votes.update({county:0})
for precIndex in range(1,2749):
    precinctRow = df.iloc[precIndex]
    per_county_votes[precinctRow['County/City']] += int(precinctRow['Total Votes Cast'].replace(',', ''))
```

```python
# Check that find_sprob in Providence is working as expected
from r2b2.minerva2 import Minerva2
from r2b2.contest import Contest
from r2b2.contest import ContestType

# Get contest
contest_name = 'Virginia 2016 presidential contest'
tally = {'Hillary R. Clinton': 1981473, 'Donald J. Trump': 1769443, 'Gary Johnson': 118274, 'Evan McMullin':54054, 'Jill Stein':27638, 'All Others':33749}
reported_winner = max(tally, key=tally.get)
winner_votes = tally[reported_winner]
total_relevant = sum(tally.values())
loser_votes = total_relevant - winner_votes
margin = (winner_votes / total_relevant) - (loser_votes / total_relevant)
contest = Contest(total_relevant,
                            tally,
                            num_winners=1,
                            reported_winners=[reported_winner],
                            contest_type=ContestType.PLURALITY)

# Create audit
M = Minerva2(.1, 1.0, contest)

# Execute a round of the audit for this sample
sample = {'Hillary R. Clinton': 100, 'Donald J. Trump': 100, 'Gary Johnson': 4, 'Evan McMullin':3, 'Jill Stein':1, 'All Others':1}
stop = M.execute_round(sum(sample.values()), sample)
print(stop)

```
