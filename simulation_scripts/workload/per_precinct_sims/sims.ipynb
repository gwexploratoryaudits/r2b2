```python
import pandas as pd

fname = 'Virginia_Elections_Database__2016_President_General_Election_including_precincts.csv'
df = pd.read_csv(fname)
df
```

```python
# create a dictionary containing a list of all ballots
# where each ballot has id number 1...N, candidate, and precinct
cands = ['Hillary R. Clinton', 'Donald J. Trump', 'Gary Johnson', 'Evan McMullin', 'Jill Stein', 'All Others']
bals = {"bals": []}

index = 0
for precIndex in range(1,2749):
    precinctRow = df.iloc[precIndex]
    for cand in cands:
        votesForCand = int(precinctRow[cand].replace(',', ''))
        for i in range(votesForCand):
            bals["bals"].append({
                'id': index,
                'precinct': precinctRow['Pct'],
                'candidate': cand
            })
            index += 1
```

```python
# check things
print('total num ballots is ' + str(len(bals["bals"]))+ ' and should be 3984631')

num = 1123412
print('ballot '+str(bals["bals"][num]['id'])+' should be num '+str(num))
```

```python
# save the bals in a json file 
import json
with open('bals.json','w') as f:
    json.dump(bals, f)
```
